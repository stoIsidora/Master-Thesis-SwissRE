```python
import pandas as pd
import numpy as np
from lifelines import CoxPHFitter
from sklearn.metrics import log_loss
from sklearn.metrics import brier_score_loss
```

```python
df_train = pd.read_csv('df_train_cph0.csv')
df_test = pd.read_csv('df_test_cph0.csv')
```

```python
# CPH1
```

```python
df_train_cph1 = df_train.copy()
df_test_cph1 = df_test.copy()
```

```python
# dummy RACEA
df_train_cph1["RACEA"] = df_train_cph1["RACEA"].astype(int)
df_test_cph1["RACEA"] = df_test_cph1["RACEA"].astype(int)

df_train_cph1 = pd.get_dummies(df_train_cph1, columns=['RACEA'], prefix='RACEA', drop_first=True)  
df_test_cph1 = pd.get_dummies(df_test_cph1, columns=['RACEA'], prefix='RACEA', drop_first=True) 

dummy_cols = [col for col in df_train_cph1.columns if col.startswith("RACEA_")]
df_train_cph1[dummy_cols] = df_train_cph1[dummy_cols].astype(int)
df_test_cph1[dummy_cols] = df_test_cph1[dummy_cols].astype(int)
```

```python
# dummy REGION
df_train_cph1["REGION"] = df_train_cph1["REGION"].astype(int)
df_test_cph1["REGION"] = df_test_cph1["REGION"].astype(int)

df_train_cph1 = pd.get_dummies(df_train_cph1, columns=['REGION'], prefix='REGION', drop_first=True)  
df_test_cph1 = pd.get_dummies(df_test_cph1, columns=['REGION'], prefix='REGION', drop_first=True) 

dummy_cols = [col for col in df_train_cph1.columns if col.startswith("REGION_")]
df_train_cph1[dummy_cols] = df_train_cph1[dummy_cols].astype(int)
df_test_cph1[dummy_cols] = df_test_cph1[dummy_cols].astype(int)
```

```python
# dummy MARST
df_train_cph1["MARST"] = df_train_cph1["MARST"].astype(int)
df_test_cph1["MARST"] = df_test_cph1["MARST"].astype(int)

df_train_cph1 = pd.get_dummies(df_train_cph1, columns=['MARST'], prefix='MARST', drop_first=True)  
df_test_cph1 = pd.get_dummies(df_test_cph1, columns=['MARST'], prefix='MARST', drop_first=True) 

dummy_cols = [col for col in df_train_cph1.columns if col.startswith("MARST_")]
df_train_cph1[dummy_cols] = df_train_cph1[dummy_cols].astype(int)
df_test_cph1[dummy_cols] = df_test_cph1[dummy_cols].astype(int)
```

```python
# dummy HSTATYR
df_train_cph1["HSTATYR"] = df_train_cph1["HSTATYR"].astype(int)
df_test_cph1["HSTATYR"] = df_test_cph1["HSTATYR"].astype(int)

df_train_cph1 = pd.get_dummies(df_train_cph1, columns=['HSTATYR'], prefix='HSTATYR', drop_first=True)  
df_test_cph1 = pd.get_dummies(df_test_cph1, columns=['HSTATYR'], prefix='HSTATYR', drop_first=True) 

dummy_cols = [col for col in df_train_cph1.columns if col.startswith("HSTATYR_")]
df_train_cph1[dummy_cols] = df_train_cph1[dummy_cols].astype(int)
df_test_cph1[dummy_cols] = df_test_cph1[dummy_cols].astype(int)
```

```python
# dummy SMOKESTATUS2
df_train_cph1["SMOKESTATUS2"] = df_train_cph1["SMOKESTATUS2"].astype(int)
df_test_cph1["SMOKESTATUS2"] = df_test_cph1["SMOKESTATUS2"].astype(int)

df_train_cph1 = pd.get_dummies(df_train_cph1, columns=['SMOKESTATUS2'], prefix='SMOKESTATUS2', drop_first=True)  
df_test_cph1 = pd.get_dummies(df_test_cph1, columns=['SMOKESTATUS2'], prefix='SMOKESTATUS2', drop_first=True) 

dummy_cols = [col for col in df_train_cph1.columns if col.startswith("SMOKESTATUS2_")]
df_train_cph1[dummy_cols] = df_train_cph1[dummy_cols].astype(int)
df_test_cph1[dummy_cols] = df_test_cph1[dummy_cols].astype(int)
```

```python
# dummy ALCSTAT1
df_train_cph1["ALCSTAT1"] = df_train_cph1["ALCSTAT1"].astype(int)
df_test_cph1["ALCSTAT1"] = df_test_cph1["ALCSTAT1"].astype(int)

df_train_cph1 = pd.get_dummies(df_train_cph1, columns=['ALCSTAT1'], prefix='ALCSTAT1', drop_first=True)  
df_test_cph1 = pd.get_dummies(df_test_cph1, columns=['ALCSTAT1'], prefix='ALCSTAT1', drop_first=True) 

dummy_cols = [col for col in df_train_cph1.columns if col.startswith("ALCSTAT1_")]
df_train_cph1[dummy_cols] = df_train_cph1[dummy_cols].astype(int)
df_test_cph1[dummy_cols] = df_test_cph1[dummy_cols].astype(int)
```

```python
cph = CoxPHFitter()


cph.fit(df_train_cph1, duration_col='TIMETOEVENT', event_col='MORTSTAT')

cph.print_summary()
```

```python
print(f'APLL_is: {cph.score(df_train_cph1, scoring_method="log_likelihood"):.4f}')
print(f'APLL_os: {cph.score(df_test_cph1, scoring_method="log_likelihood"):.4f}')
print(f'APLL_is: {cph.score(df_train_cph1, scoring_method="concordance_index"):.4f}')
print(f'APLL_os: {cph.score(df_test_cph1, scoring_method="concordance_index"):.4f}')
```

```python
# predict the survival function for each individual
survival_probs_16_train = cph.predict_survival_function(df_train_cph1)
survival_probs_16_test = cph.predict_survival_function(df_test_cph1)

# extract 3y survival probabilities
survival_prob_3y_train = survival_probs_16_train.iloc[3] 
survival_prob_3y_test = survival_probs_16_test.iloc[3] 

# get 3-year event status based on TIMETOEVENT and MORTSTAT
status_3y_train = np.where((df_train_cph1["MORTSTAT"] == 1) & (df_train_cph1["TIMETOEVENT"] <= 3), 0, 1)
status_3y_test = np.where((df_test_cph1["MORTSTAT"] == 1) & (df_test_cph1["TIMETOEVENT"] <= 3), 0, 1)

print(f'log-loss is: {log_loss(status_3y_train, survival_prob_3y_train):.4f}')
print(f'log-loss os: {log_loss(status_3y_test, survival_prob_3y_test):.4f}')
print(f'BS is: {brier_score_loss(status_3y_train, survival_prob_3y_train):.4f}')
print(f'BS os: {brier_score_loss(status_3y_test, survival_prob_3y_test):.4f}')
```
